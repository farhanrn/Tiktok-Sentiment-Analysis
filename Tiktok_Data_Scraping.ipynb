```python
from google_play_scraper import reviews_all, Sort

# Scrape reviews for TikTok app in Indonesian (ID) from Indonesia
try:
    tiktok_reviews = reviews_all(
        'com.ss.android.ugc.trill',  # TikTok app ID
        lang='id',                   # Set to Indonesian language
        country='id',                # Set to Indonesia
        sort=Sort.MOST_RELEVANT,      # Sort by most relevant reviews
        count= 1000000
    )

    # Print or process the reviews
    print(tiktok_reviews[:5])  # Display first 5 reviews as a sample

except Exception as e:
    print("An error occurred:", e)
```

```python
# Menyimpan ulasan dalam file CSV
import csv

with open('tiktok_review.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review'])  # Menulis header kolom
    for review in tiktok_reviews:
        writer.writerow([review['content']])  # Menulis konten ulasan ke dalam file CSV
```
